```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("Resources/charity_data.csv")
application_df.head()
```

```python
# pre-processing - same as original 

df = application_df.drop(["EIN", "NAME"], axis=1)

app_types = df['APPLICATION_TYPE'].value_counts()
application_types_to_replace = list(app_types[app_types<500].index)
for app in application_types_to_replace:
    df['APPLICATION_TYPE'] = df['APPLICATION_TYPE'].replace(app,"Other")
    
class_types = df['CLASSIFICATION'].value_counts()
classifications_to_replace = list(class_types[class_types<100].index)
for cls in classifications_to_replace:
    df['CLASSIFICATION'] = df['CLASSIFICATION'].replace(cls,"Other")

num_df = pd.get_dummies(df)

y = num_df["IS_SUCCESSFUL"]
X = num_df.drop(["IS_SUCCESSFUL"],axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)

scaler = StandardScaler()
X_scaler = scaler.fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Optimization 1: starting by adding more epochs and more neurons in hidden layers

number_input_features = len(X_train_scaled)
nn = tf.keras.models.Sequential()

nn.add(tf.keras.layers.Dense(units=12, input_shape=(X_train_scaled.shape[1],), activation="relu", input_dim=number_input_features))
nn.add(tf.keras.layers.Dense(units=12, input_shape=(X_train_scaled.shape[1],), activation="relu"))
nn.add(tf.keras.layers.Dense(units=1, input_shape=(X_train_scaled.shape[1],), activation="sigmoid"))

nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

fit_model = nn.fit(X_train_scaled, y_train, epochs=300)

model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Optimization 2: change activation functions in hidden layers

number_input_features = len(X_train_scaled)
nn = tf.keras.models.Sequential()

nn.add(tf.keras.layers.Dense(units=12, input_shape=(X_train_scaled.shape[1],), activation="selu", input_dim=number_input_features))
nn.add(tf.keras.layers.Dense(units=12, input_shape=(X_train_scaled.shape[1],), activation="selu"))
nn.add(tf.keras.layers.Dense(units=1, input_shape=(X_train_scaled.shape[1],), activation="sigmoid"))

nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

fit_model = nn.fit(X_train_scaled, y_train, epochs=300)

model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Optimization 3: add 3rd hidden layer, revert back to 200 epochs

number_input_features = len(X_train_scaled)
nn = tf.keras.models.Sequential()

nn.add(tf.keras.layers.Dense(units=12, input_shape=(X_train_scaled.shape[1],), activation="selu", input_dim=number_input_features))
nn.add(tf.keras.layers.Dense(units=12, input_shape=(X_train_scaled.shape[1],), activation="selu"))
nn.add(tf.keras.layers.Dense(units=16, input_shape=(X_train_scaled.shape[1],), activation="selu"))
nn.add(tf.keras.layers.Dense(units=1, input_shape=(X_train_scaled.shape[1],), activation="sigmoid"))

nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

fit_model = nn.fit(X_train_scaled, y_train, epochs=200)

model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
